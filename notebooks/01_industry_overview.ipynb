```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 
sns.set(style="whitegrid")

# 1. 
df = pd.read_csv("/Users/anniezhu/VSCode/insurance_analysis_project/data/us_insurance_premium_simulated.csv")
df.head()

# 2. 
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Year", y="Premium_Billion_USD", hue="Insurance_Type", marker="o")
plt.title(" U.S. Insurance Premium Trends by Type (20152023)", fontsize=14)
plt.ylabel("Total Premium (Billion USD)")
plt.xlabel("Year")
plt.legend(title="Insurance Type", loc="upper left")
plt.tight_layout()
plt.show()

# 3. 
summary = df.groupby("Insurance_Type").agg({
    "Premium_Billion_USD": ["mean", "max"]
}).round(2)

summary.columns = ['Average Premium', 'Max Premium']
summary = summary.sort_values("Average Premium", ascending=False)
summary
```
